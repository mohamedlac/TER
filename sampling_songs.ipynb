```python
import pandas as pd
import numpy as np
import scipy.io
import sys
```

```python
songs = pd.read_csv("data/songs_clustered.csv",names = ["frequency","popularity","mean","median","cluster","song_id"])
songs.head(5)
```

```python
dataset = pd.read_csv("data/songsDataset.csv",header = 0)
dataset.head(5)
len(dataset)
```

```python
popular_songs = songs.query("cluster == 2")
```

```python
len(popular_songs)
```

```python
popular_songs_id =list(popular_songs["song_id"])
users = dataset[dataset["'songID'"].isin(popular_songs_id)]

users = users.rename(columns={"'songID'":"song_id","'userID'":"user_id","'rating'":"rating"})
```

```python
len(users.song_id.unique())
```

```python
users_id = list(users["user_id"])

data_sample = dataset[dataset["'userID'"].isin(users_id)]
data_sample.head(5)
```

```python
data_sample = data_sample.rename(columns={"'songID'":"song_id","'userID'":"user_id","'rating'":"rating"})
```

```python
sample_songs = list(data_sample["song_id"])
```

```python
sample_clusters = songs[songs["song_id"].isin(sample_songs)]
sample_clusters.head(5)
```

```python
sample_clusters.groupby("cluster").describe()
```

```python
cluster_2 = sample_clusters.query("cluster ==2")
cluster_1 = sample_clusters.query("cluster ==1")
cluster_0 = sample_clusters.query("cluster ==0")
```

```python
cluster_2_subset = cluster_2
cluster_0_subset = cluster_0.sample(frac=0.1, random_state = 2)
cluster_1_subset = cluster_1.sample(n=2554,random_state =3)
```

```python
len(cluster_0_subset)+ len(cluster_1_subset)+len(cluster_2_subset)
```

```python
final_songs_sample = pd.concat([cluster_0_subset,cluster_1_subset,cluster_2_subset])
final_songs_sample.head(10)
```

```python
final_songs_id = list(final_songs_sample['song_id'])

final_data_sample = dataset[dataset["'songID'"].isin(final_songs_id)]
```

```python
final_data_sample = final_data_sample.rename(columns={"'songID'":"song_id","'userID'":"user_id","'rating'":"rating"})
```

```python
user_song_mat = final_data_sample.pivot(index ='user_id',columns='song_id',values="rating")
```

```python
user_song_mat2 = user_song_mat.sample(n=5000,random_state =3)
```

```python
user_song_mat2.head(5)
```

```python
user_song_mat2 = user_song_mat2.fillna(0)
```

```python
user_song_mat2.to_csv("data/user_song_matrix.csv",header=False)
```

```python
matrix = np.array(user_song_mat2)
```

```python
matrix.shape
```

```python
scipy.io.savemat("data/user_song_matrix", {'csvmatrix':matrix})
```
